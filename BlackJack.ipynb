```python
import random

suits = ('Hearts', 'Kings', 'Spades', 'Clubs')
nums = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 
        'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 
         'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
playing = True
```

```python
class Card():
    
    def __init__(self,suit,num):
        self.suit = suit
        self.num = num
        
        
    def __str__(self):
        return self.num + " of " + self.suit
```

```python
class Deck():
    
    def __init__(self):
        self.card_deck = []
        for suit in suits:
            for num in nums:
                card_created = Card(suit,num)
                self.card_deck.append(card_created)
                
    def __str__(self):
        n = 0
        deck = ''
        for crd in self.card_deck:
            deck += '\n'+ crd.__str__()
            n = n+1
        return "The deck has: "+deck
            
                
    def shuffle_deck(self):
        random.shuffle(self.card_deck)
        
    def deal(self):
        popped_card = self.card_deck.pop()
        return popped_card
```

```python
class Hand():
    
    def __init__(self):
        self.card_deck2 = []
        self.value = 0
        self.aces = 0
        
    def add_card(self,card):
        self.card_deck2.append(card)
        self.value += values[card.num]
        
        #no. of aces:
        if card.num == 'Ace':
            self.aces += 1
    
    def adjust_for_aces(self):
        
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
class Chips():
    def __init__(self,total = 100):
        self.total = total
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    
    while True:
        
        try:
            chips.bet = int(input("How many chips do you want to bet: "))
        except:
            print("Sorry please provide an integer")
        else:
            if chips.bet>chips.total:
                print("sorry, you do not have enough chips! You have: {}".format(chips.total))
            else:
                break
    
```

```python
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_aces()
```

```python
def hit_or_stand(deck,hand):
    global playing
    
    while True:
        inp = input("Hit or Stand? Enter h/s: ")
        
        if inp[0].lower() == 'h':
            hit(deck,hand)
            
        elif inp[0].lower() == 's':
            print("Player Stands, Dealser's Turn!!")
            playing = False
            
        else:
            print("Sorry didn't understand the input, please enter h/s: ")
            continue
        
        break
```

```python
def show_some(player,dealer):
    print("\nDealer's Hand:")
    print(" <card hidden>")
    print('',dealer.card_deck2[1])  
    print("\nPlayer's Hand:", *player.card_deck2, sep='\n ')
    
def show_all(player,dealer):
    print("\nDealer's Hand:", *dealer.card_deck2, sep='\n ')
    print("Dealer's Hand =",dealer.value)
    print("\nPlayer's Hand:", *player.card_deck2, sep='\n ')
    print("Player's Hand =",player.value)
```

```python
def player_busts(player,dealer,chips):
    print("BUST PLAYER")
    chips.lose_bet()

def player_wins(player,dealer,chips):
    print("PLAYER WINS")
    chips.win_bet()

def dealer_busts(player,dealer,chips):
    print("BUST DEALER, PLAYER WINS")
    chips.lose_bet()
    
def dealer_wins(player,dealer,chips):
    print("DEALER WINS")
    chips.win_bet()
    
def push(player,dealer):
    print("Dealer and Player TIE!! PUSH")
```
